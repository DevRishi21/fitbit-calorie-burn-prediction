```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

# Load Cleaned Dataset

```python
cleaned_df = pd.read_csv("../Fitbit/data/fitbit_cleaned.csv")
cleaned_df.head()
```

```python
cleaned_df.info()
```

# Train_Test_Split

```python
X = cleaned_df.drop('Calories_Burned', axis = 1)
Y = cleaned_df['Calories_Burned']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.20,
    random_state=42
)
```

## Fit and Transform the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

# The scaler calculates the mean and standard deviation for each feature using fit()
X_train_scaled = scaler.fit_transform(X_train)

# Each training value is converted to its standardized value using transform()
X_test_scaled = scaler.transform(X_test)

# converting it to Dataframe
X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns,
    index=X_train.index
)

X_test_scaled = pd.DataFrame(
    X_test_scaled,
    columns=X_test.columns,
    index=X_test.index
)

X_train_scaled.tail()
```

# Helper function for Model Evaluation

```python
import numpy as np

def eval_model(model, x_train, y_train, x_test, y_test, model_name):
    model.fit(x_train, y_train)
    y_predict = model.predict(x_test)

    # Model Evaluation
    mae = mean_absolute_error(y_test, y_predict)
    rmse = np.sqrt(mean_squared_error(y_test, y_predict))
    r2 = r2_score(y_test, y_predict)

    print(f"{model_name}")
    print("-" * 35)
    print(f"MAE  : {mae:.4f}")
    print(f"RMSE : {rmse:.4f}")
    print(f"R² Score : {r2:.4f}")
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

lr_model = LinearRegression()

eval_model(lr_model, X_train, Y_train, X_test, Y_test, "Linear Regression")
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)

eval_model(ridge_model, X_train, Y_train, X_test, Y_test, "Ridge Regression")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

eval_model(rf_model, X_train, Y_train, X_test, Y_test, "Random Forest")
```

# KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor(n_neighbors=5)

eval_model(knn_model, X_train_scaled, Y_train, X_test_scaled, Y_test, "KNN Regressor")
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor()

print("Unscaled Data")
eval_model(dt_model, X_train, Y_train, X_test, Y_test, "Decision Tree")

print("-"*30)
print("Scaled Data")
eval_model(dt_model, X_train_scaled, Y_train, X_test_scaled, Y_test, "Decision Tree")
```

# SVR

```python
from sklearn.svm import SVR

svr_model = SVR()

print("Unscaled Data")
eval_model(svr_model, X_train, Y_train, X_test, Y_test, "Support Vector Regression")

print("-"*30)
print("Scaled Data")
eval_model(svr_model, X_train_scaled, Y_train, X_test_scaled, Y_test, "Support Vector Regression")
```
